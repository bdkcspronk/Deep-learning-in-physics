```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
normal_df = pd.read_csv('train_losses.csv')
curriculum_df = pd.read_csv('curriculum_train_losses.csv')

marker_size = 250

plt.figure(figsize=(8,6))
plt.scatter(normal_df['epoch'], normal_df['loss'], marker='o', s=marker_size, facecolors='none', color='0.5', label='Normal')
plt.scatter(curriculum_df['epoch'], curriculum_df['loss'], marker='x', s=marker_size, color='0.5', label='Curriculum')

# Add shaded regions for each stage
plt.axvspan(1, 5.5, color="#2ca02c", alpha=0.15, label='High Confidence Stage')
plt.axvspan(5.5, 10.5, color="#26a399", alpha=0.15, label='High+Medium Confidence Stage')
plt.axvspan(10.5, 20.5, color="#4732a5", alpha=0.10, label='Full Dataset Stage')

plt.xlabel('Epoch', fontsize=20)
plt.ylabel('Training Loss', fontsize=20)
plt.title('Normal vs. Curriculum Learning', fontsize=22)
plt.tick_params(axis='both', labelsize=16)
plt.yticks([0.26, 0.29, 0.32, 0.35, 0.38, 0.41, 0.44, 0.47])
plt.xticks([1, 6, 11, 20])
ticks = list(plt.xticks()[0]) + [3.5, 8.5, 15.5]
labels = [item.get_text() for item in plt.gca().get_xticklabels()] + ['high', 'medium', 'full']
plt.xticks(ticks, labels, rotation=45)
plt.xticks(ticks, labels)
plt.legend(fontsize=16)
plt.grid(True, alpha=0.8, linestyle='--')
plt.tight_layout()
plt.savefig('training_loss_comparison.png', dpi=150)
plt.show()
```
